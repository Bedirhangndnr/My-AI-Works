**Boxplot
Veri Seti Hikayesi**

total_bill: 
yemeğin toplam fiyatı (bahşiş ve vergi dahil)

tip: bahşiş

sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)

smoker: grupta sigara içen var mı? (0=No, 1=Yes)

day: gün (3=Thur, 4=Fri, 5=Sat, 6=Sun)

time: ne zaman? (0=Day, 1=Night)

size: grupta kaç kişi var?

```python
import seaborn as sns
diamonds = sns.load_dataset('tips')
df = diamonds.copy()
df.head()
```

```python
#amacımızı hatırlayalım
#verileri sorguluyoruz. Verileri tanımalıyız. bu öneriler önemlidir. iş yaparken farkında olunmalı
df.describe().T
```

```python
df["sex"].value_counts()
```

```python
df["smoker"].value_counts()
```

```python
df["day"].value_counts()
```

```python
sns.boxplot(x=df["total_bill"])
```

```python
#boxplot (kutu grafiği):Kutu bıyık grafiği—kutu grafiği olarak daadlandırılır—bir veri
#setinin beş sayılı özetini gösterir.Beş sayılı özet minimum, birinci çeyrek,
#medyan, üçüncü çeyrek ve maksimumdur. Aykırı durum işlemede çok kullanılır.
#khan academy makalesi:
#https://tr.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/box-plot-review
```

```python
#istersek dikeyde elde edebiliriz.
sns.boxplot(x=df["total_bill"],orient="v")
#orient= yönlendirme
```

```python
#mesela burada biz cuma ve çarşamba daha fazla aile gelmiş diyebilir miyiz? 
#bunun için toplu kişi sayısı veya çift sayısı verisi varsa kullanabiliriz.
sns.boxplot(x="day",y="tip",hue="sex",data=df);
```

```python
sns.catplot(y="total_bill",kind="violin",data=df);
```

Violin grafiğimize bir boyut ekleyelim

```python
sns.catplot(x="day",y="total_bill",hue="sex",kind="violin",data=df);
```

**KORELASYON GRAFİKLERİ**

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

KORELASYON ÇARPRAZLAMALAR

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="size",size="size",style="time",data=df)
```

**DOĞRUSAL İLİŞKİNİN GÖZLEMLENMESİ**

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lmplot(x="total_bill",y="tip",data=df)
#l = lineerin kısaltılmışı
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df)
```

```python
sns.lmplot(x="total_bill",y="tip",col="time",hue="smoker",data=df)
```

```python
sns.lmplot(x="total_bill",y="tip",col="time",hue="smoker",row="sex",data=df)
```
