total_bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)

tip: bahşiş

sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)

smoker: grupta sigara içen var mı? (0=No, 1=Yes)

day: gün (3=Thur, 4=Fri, 5=Sat, 6=Sun)

time: ne zaman? (0=Day, 1=Night)

size: grupta kaç kişi var?

# FACED GRİDDE KALDIK:

```python
sns.distplot(df.total_bill,bins=30,hist=False)
```

```python
sns.FacetGrid(df,hue="day",height=7,xlim=(0,60)).map(sns.kdeplot,"total_bill",shade=True).add_legend()
```

```python
import matplotlib.pyplot as plt
sns.FacetGrid(df, col="time", row="smoker",height=3).map(plt.hist,"total_bill",color="b").add_legend()
```

```python
sns.FacetGrid(df,col="time",row="smoker",height=4).map(plt.scatter,"total_bill","tip",edgecolor="w")
```

```python
#sns.FacetGrid(df,col="day",height=4,aspect=.5).map(plt.hist,"total_bill",bins=bins)
g = sns.FacetGrid(df, col="day", height=4, aspect=.5)
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(df, col="smoker", col_order=["Yes", "No"])
g = g.map(plt.hist, "total_bill", color="m")
```

```python
kws=dict(s=50,linewidth=.4,edgecolor="y")
sns.FacetGrid(df,col="sex",hue="time",palette="Set1",
             hue_order=["Dinner", "Lunch"],height=5).map(plt.scatter,
                                               "total_bill","tip",**kws).add_legend()
```

```python
kws=dict(s=50,linewidth=.4,edgecolor="y")
sns.FacetGrid(df,col="sex",
              hue="time",hue_kws=dict(marker=["v","*"]),
              palette="Set1",
             hue_order=["Dinner", "Lunch"],
              height=5).map(plt.scatter,"total_bill","size",**kws).add_legend()
```

```python
sns.catplot(x="day",y="tip",hue="sex",kind="point",data=df)
```
