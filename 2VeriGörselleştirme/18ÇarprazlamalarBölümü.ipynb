```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
df.head()
#çarprazlanma= veri seti içerisinde yer alan degişkenlerin birlikte değerlendirlmesi denir.
#artık daha somut yorumlar yapacağız.......
#ÖNEMLİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİİ!!!!!!!!!!!!!!!!!!!!!!
```

```python
sns.catplot(x = "cut", y = "price", data = df);
```

```python
#bir boyut daha ekleyelim
sns.barplot(x="cut",y="price",hue="color",data=df);
```
