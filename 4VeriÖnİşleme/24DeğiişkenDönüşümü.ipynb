```python
import pandas as pd
df_one_hot=pd.get_dummies(df,columns=["sex"],prefix=["sexxxx"])
#prefix= oluşturulacak olan değişkenlerin ön isimlendirmesi ne olacak?
df_one_hot.head()
#sondaki 2 değişkene odaklanalım.
```

```python
# Burada ki "Dummy Değişken Tuzağı" şu; sex_male ve sex_female aynı bilgiyi taşıyor.
# Yani biri ile ötekini elde edebiliyoruz. bu bir problem.
```

```python
#mesela bu işilemi birde day değişkenine yapalım.
df_one_hot=pd.get_dummies(df,columns=["day"],prefix=["dayyyy"]) # dediğimizdeüstteki sorun olmayacak ama 
df_one_hot.head()                                               # çünkü 2den fazla değişken var
```
