```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3})

df = df.astype(float)
df
#standardizasyon veri setinin taşıdığı bilgiyi bozmadan belirli br standarda getirir
#dönüştürme ise veri setinin taşımış olduğu bilgiyi bazen bozar,bazen temsil şeklini değiştirir
```

**SÜREKLİ DEĞİŞKENİ KATEGORİK DEĞİŞKENE ÇEVİRMEK**

```python
df.head()
```

```python
dff=df.select_dtypes(include=["float64","int64"]) 
```

```python
est=preprocessing.KBinsDiscretizer(n_bins=[20,2,3],encode="ordinal",strategy="quantile").fit(dff)
# n_bins = [3,2,2] olarak girdiğimizde buradaki 3, 3 e ayır anlamında.dff 'in ilk columnu
# yani total_bill değişkeninin değerlerini büyüklüklerine göre  (0,1,2) olarak atıyor. 2 ler 
# ise tip ve size değişkenlerinin değerlerini 0 veya 1 olarak atıyor. Yani büyüklüklerine göre 
# kategorik değişkene çevirmiş olduk.
est.transform(dff)[0:10]
```
